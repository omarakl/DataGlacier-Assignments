```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_1 = pd.read_csv('Customer_ID.csv')
data_2 = pd.read_csv('Transaction_ID.csv')
cab_data = pd.read_csv('Cab_Data.csv')
```

```python
# merging 'customer_data' and 'transaction_data'
customer_data = pd.merge(data_1, data_2, how='inner')
```

```python
customer_data.dropna(inplace = True) # dropping NaN values
```

```python
customer_data.drop_duplicates()  # drop duplicates
```

```python
# Showing the number of males and females
sns.countplot(x='Gender', data=customer_data)
```

```python
plt.figure(figsize=(13, 6))
plt.xticks(rotation=45)
ax = sns.countplot(x='Age', data=customer_data, palette='viridis', saturation=0.75, width=0.9)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=7, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

```python
sns.countplot(x='Company', data=cab_data)

# Add labels and title
plt.xlabel('Comapany')
plt.ylabel('Count')
plt.title('Number users of each cab company')

# Show the plot
plt.show()
```
